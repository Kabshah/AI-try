## BIVARIATE ANALYSIS

```python
gender_counts = data['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightpink'])
plt.title("Bars showing number of Females and Males")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
#Show exact numbers on top of bars
for i, val in enumerate(gender_counts.values):
    plt.text(i, val + 20, str(val), ha='center', fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
data.boxplot(column="Review Rating", by="Gender")
plt.xlabel("Gender")
plt.ylabel("Review Rating")
plt.title("Review Rating Distribution by Gender")
plt.show()
```

## UNIVARIATE ANALYSIS

```python
# Create age groups (customize bins as needed)
bins = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['<18', '18-24', '25-34', '35-44', '45-54', '55-64', '65+']
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)

# Calculate age group distribution
age_distribution = data['Age Group'].value_counts().sort_index()
# 2. Visualize the distribution
plt.figure(figsize=(12, 6))
ax = age_distribution.plot(kind='bar', color='teal')
plt.title('Customer Distribution by Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)

# Add percentage labels
total_customers = len(data)
for i, count in enumerate(age_distribution):
    percentage = (count / total_customers) * 100
    ax.text(i, count + 20, f'{percentage:.1f}%', ha='center')
plt.show()
```
